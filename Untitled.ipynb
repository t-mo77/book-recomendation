```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books=pd.read_csv(path+"/books.csv")    
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("Column")
plt.ylabel("Missing Value")
plt.show()
```

```python
import numpy as np  
books["missing_description"] = np.where(books["description"].isnull(),1,0) 
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]
correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Spearman Correlation"})
heatmap.set_title("Spearman Correlation Matrix")
plt.show()
```

```python
book_missing=books[~(books["description"].isnull()) & ~(books["num_pages"].isnull()) & ~(books["average_rating"].isnull()) & ~ (books["published_year"].isnull())]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(1,4), "description"] 
```

```python
book_missing.loc[book_missing["words_in_description"].between(5,14), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(15,24), "description"]
```

```python
book_missing_25_words=book_missing.loc[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"]=(np.where(book_missing_25_words["subtitle"].isna(),book_missing_25_words["title"],book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1)))
```

```python
book_missing_25_words   
```

```python
book_missing_25_words["tagged_description"]=book_missing_25_words[["isbn13","description","title_and_subtitle"]].astype(str).agg(": ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(columns=["subtitle","missing_description","words_in_description","age_of_book"],axis=1)
    .to_csv("books_cleaned.csv",index=False)
)
```
